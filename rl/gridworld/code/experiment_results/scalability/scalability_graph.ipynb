```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'state_action_space_scalability.csv'
df = pd.read_csv(file_path)

# Plotting the graph
plt.figure(figsize=(8, 6))
plt.plot(df['Number of Agents'], df['State-Action Space Size'], marker='o', color='r', label="State-Action Space Complexity")

# Add labels for each data point
for i in range(len(df)):
    plt.text(df['Number of Agents'][i], df['State-Action Space Size'][i], f"({df['Number of Agents'][i]}, {df['State-Action Space Size'][i]})", 
             fontsize=9, ha='right')

plt.title('Scalability Issue in MARL: Growth of State-Action Space with Number of Agents')
plt.xlabel('Number of Agents')
plt.ylabel('State-Action Space Size')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
