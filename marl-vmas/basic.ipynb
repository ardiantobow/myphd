```python
import vmas

# Create a VMAS environment with a specific scenario
env = vmas.make_env(
    scenario="waterfall",  # Scenario name or a custom scenario class
    num_envs=32,          # Number of parallel environments
    device="cpu",         # Compute device ("cpu" or "cuda")
    continuous_actions=True, # Whether the action space is continuous
    max_steps=None,       # Maximum number of steps (None for infinite horizon)
    seed=None,            # Random seed for reproducibility
    n_agents=3            # Number of agents in the environment
)

# Reset the environment and get initial observations
obs = env.reset()

# Step through the environment with random actions
for _ in range(10):
    actions = env.get_random_actions()
    obs, rewards, dones, info = env.step(actions)
    print("Observations:", obs)
    print("Rewards:", rewards)
    print("Dones:", dones)
    print("Info:", info)
```
